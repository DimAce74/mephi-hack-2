```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv('/kaggle/input/mephikack2/cleared.csv')
df.info()
```

```python
df.head()
```

```python
def concat_text(row):
    
    text = str(row['doc_text'])
    
    if not pd.isna(str(row['image2text'])):
        text += str(row['image2text'])

    if not pd.isna(str(row['speech2text'])):
        text += str(row['speech2text'])

    return text.lower()

df['text'] = df.apply(concat_text, axis=1)

df.head()
```

```python
df = df.drop(columns=['doc_text', 'image2text', 'speech2text'])
```

```python
df. describe()
```

```python
df['class'] = df['—Ç–µ–º—ã']
df = df.drop(columns=['—Ç–µ–º—ã'])
```

```python
df.info()
```

```python
x_for_embedding = df['text']
vectorizer = TfidfVectorizer(max_features=30000, use_idf=True)
X = vectorizer.fit_transform(x_for_embedding)
```

```python
import re
import ast


def parse_labels(label):
    if isinstance(label, str):
        # –ï—Å–ª–∏ —ç—Ç–æ —Å—Ç—Ä–æ–∫–∞, –ø—ã—Ç–∞–µ–º—Å—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –≤ —Å–ø–∏—Å–æ–∫
        try:
            # –£–¥–∞–ª—è–µ–º –ª–∏—à–Ω–∏–µ –ø—Ä–æ–±–µ–ª—ã –∏ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å—Ç—Ä–æ–∫—É –≤ —Å–ø–∏—Å–æ–∫
            label = ast.literal_eval(label)
        except (ValueError, SyntaxError):
            # –ï—Å–ª–∏ –Ω–µ —É–¥–∞–ª–æ—Å—å, –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –∫–∞–∫ —Å—Ç—Ä–æ–∫—É
            label = re.findall(r'[–∞-—è–ê-–Øa-zA-Z\s]+', label)
            label = [l.strip() for l in label if l.strip()]
    elif isinstance(label, list):
        # –ï—Å–ª–∏ —ç—Ç–æ —É–∂–µ —Å–ø–∏—Å–æ–∫, –æ—Å—Ç–∞–≤–ª—è–µ–º –∫–∞–∫ –µ—Å—Ç—å
        label = [str(l).strip() for l in label if str(l).strip()]
    else:
        label = []
    return label

y = df['class']
y_cleaned = [parse_labels(label) for label in y]
y_cleaned
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
y_transformed = mlb.fit_transform(y_cleaned)

# –ù–∞–∑–Ω–∞—á–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–æ–≤ –∫–ª–∞—Å—Å–∞–º: —Å–ø–æ—Ä—Ç ‚Äî 0, –ª–∏—á–Ω–∞—è –∂–∏–∑–Ω—å ‚Äî 1, —Å–æ—Ü—Å–µ—Ç–∏ ‚Äî 2, —Ä–µ–∫–ª–∞–º–∞ ‚Äî 3
class_mapping = {label: idx for idx, label in enumerate(mlb.classes_)}
y_transformed
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, hamming_loss

def transformer_block(inputs, head_size, num_heads, ff_dim, dropout=0.1):
    # Multi-Head Attention
    x = layers.MultiHeadAttention(
        key_dim=head_size, num_heads=num_heads, dropout=dropout
    )(inputs, inputs)
    x = layers.Dropout(dropout)(x)
    x = layers.LayerNormalization(epsilon=1e-6)(x + inputs)  # Residual connection

    # Feed-Forward Network
    ffn = layers.Dense(ff_dim, activation="relu")(x)
    ffn = layers.Dense(inputs.shape[-1])(ffn)
    ffn = layers.Dropout(dropout)(ffn)
    x = layers.LayerNormalization(epsilon=1e-6)(ffn + x)  # Residual connection
    return x

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
def build_transformer_model(input_dim, num_classes, head_size=64, num_heads=4, ff_dim=128, dropout=0.1):
    inputs = layers.Input(shape=(input_dim,))

    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –≤ —Ñ–æ—Ä–º–∞—Ç [batch_size, sequence_length, feature_dim]
    x = layers.Reshape((1, input_dim))(inputs)  # –î–æ–±–∞–≤–ª—è–µ–º —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –¥–ª—è attention

    # –¢—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä–Ω—ã–π –±–ª–æ–∫
    x = transformer_block(x, head_size, num_heads, ff_dim, dropout)

    # Pooling –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –µ–¥–∏–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞
    x = layers.GlobalAveragePooling1D()(x)

    # –í—ã—Ö–æ–¥–Ω–æ–π —Å–ª–æ–π –¥–ª—è –º–Ω–æ–≥–æ–º–µ—Ç–æ—á–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    outputs = layers.Dense(num_classes, activation="sigmoid")(x)

    model = models.Model(inputs, outputs)
    return model

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –º–æ–¥–µ–ª–∏
input_dim = X.shape[1]  # –ù–∞–ø—Ä–∏–º–µ—Ä, 100 (–∑–∞–≤–∏—Å–∏—Ç –æ—Ç max_features –≤ TfidfVectorizer)
num_classes = y_transformed.shape[1]  # –ù–∞–ø—Ä–∏–º–µ—Ä, 4 (—Å–ø–æ—Ä—Ç, –ª–∏—á–Ω–∞—è –∂–∏–∑–Ω—å, —Å–æ—Ü—Å–µ—Ç–∏, —Ä–µ–∫–ª–∞–º–∞)

# –°–æ–∑–¥–∞—ë–º –º–æ–¥–µ–ª—å
model = build_transformer_model(
    input_dim=input_dim,
    num_classes=num_classes,
    head_size=128,      # –†–∞–∑–º–µ—Ä –∫–ª—é—á–∞ –¥–ª—è attention
    num_heads=16,       # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–æ–ª–æ–≤ attention
    ff_dim=256,        # –†–∞–∑–º–µ—Ä Feed-Forward —Å–ª–æ—è
    dropout=0.1        # Dropout –¥–ª—è —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏
)

# –ö–æ–º–ø–∏–ª—è—Ü–∏—è –º–æ–¥–µ–ª–∏
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="binary_crossentropy",  # –î–ª—è –º–Ω–æ–≥–æ–º–µ—Ç–æ—á–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    metrics=["accuracy"]
)



# –ü—Ä–∏–º–µ—Ä –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# new_texts = ["—Ç–µ–∫—Å—Ç –ø—Ä–æ —Å–ø–æ—Ä—Ç –∏ —Å–æ—Ü—Å–µ—Ç–∏", "—Ä–µ–∫–ª–∞–º–∞ –∏ –ª–∏—á–Ω–∞—è –∏—Å—Ç–æ—Ä–∏—è"]
# X_new = vectorizer.transform(new_texts).toarray()  # –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è, —á—Ç–æ vectorizer —É–∂–µ –æ–±—É—á–µ–Ω
# y_new_pred = model.predict(X_new)
# y_new_pred_binary = (y_new_pred > 0.5).astype(int)
# predicted_labels = mlb.inverse_transform(y_new_pred_binary)
# print("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –º–µ—Ç–∫–∏:", predicted_labels)
```

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º X –≤ numpy, –µ—Å–ª–∏ –æ–Ω sparse (–æ—Ç TfidfVectorizer)
X = X.toarray() if hasattr(X, "toarray") else X
y_transformed = np.array(y_transformed)

# –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â–∏–µ –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã–µ
X_train, X_val, y_train, y_val = train_test_split(
    X, y_transformed, test_size=0.3, random_state=42
)

# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
history = model.fit(
    X_train,
    y_train,
    batch_size=16,
    epochs=20,
    verbose=1
)

# –û—Ü–µ–Ω–∫–∞ –º–æ–¥–µ–ª–∏
y_pred = model.predict(X_val)
y_pred_binary = (y_pred > 0.5).astype(int)  # –ë–∏–Ω–∞—Ä–∏–∑–∞—Ü–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π

# –ú–µ—Ç—Ä–∏–∫–∏
f1 = f1_score(y_val, y_pred_binary, average="micro")
hamming = hamming_loss(y_val, y_pred_binary)
print("F1 Score (micro):", f1)
print("Hamming Loss:", hamming)
```
